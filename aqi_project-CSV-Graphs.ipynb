```python
import pandas as pd
import numpy as np 
import requests
import json
from pprint import pprint
from datetime import datetime
import matplotlib.pyplot as plt
import time
import gmaps
#import matplotlib as plt
import os
import csv
```

```python
data2021 = "aqi_data.csv"
```

```python
aqi_data = pd.read_csv(data2021)
aqi_data
```

```python
pm25_data = aqi_data.plot(kind="line", x="Date", y="PM25", grid=True, figsize=(17,8),
              title="PM2.5 over time")

plt.xlabel("Time MM/DD/YYYY")
plt.ylabel("PM units")
plt.title("PM2.5 Overtime.")


plt.savefig("Plot_Images/pm25_data.png")
```

```python
pm25_data = aqi_data.plot(kind="line", x="Date", y="PM25", grid=True, figsize=(25,12),
              title="PM2.5 over time")

plt.xlabel("Time MM/DD/YYYY")
plt.ylabel("PM units")
plt.title("PM2.5 Overtime.")
plt.savefig("Plot_Images/pm25_data2.png")
```

```python
AQI_data = aqi_data.plot(kind="line", x="Date", y="Overall AQI Value", grid=True, figsize=(25,12),
              title="AQI value over time")
plt.xlabel("Time MM/DD/YYYY")
plt.ylabel("AQI units")
plt.title("AQI Overtime.")

plt.savefig("Plot_Images/AQI_data.png")
```

```python
clean_aqi_df = aqi_data.dropna(how = 'any')
clean_aqi_df = clean_aqi_df.replace('.', None)
```

```python
clean_aqi_df["NO2"] = clean_aqi_df["NO2"].astype(float)
```

```python
NO2_data = clean_aqi_df.plot(kind="line", x="Date", y="NO2", grid=True, figsize=(25,12),
              title="NO2 over time")
plt.xlabel("Time MM/DD/YYYY")
plt.ylabel("NO2 units")
plt.title("NO2 Overtime.")
plt.show
plt.savefig("Plot_Images/NO2_data.png")
```
